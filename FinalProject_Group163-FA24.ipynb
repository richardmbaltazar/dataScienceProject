# Abstract

It is a popular idea that UCSD students hold the belief that professors with higher average grades given and lower hours worked should be prioritized during the first pass of enrollment. This perception is based on the idea that such professors and their courses are "easier" or less time-consuming, making them more desirable for students balancing demanding schedules and seeking academic success with minimal effort. However, how accurate is this assumption? Our final project for this class aimed to investigate the relationship between a professor's average grades given, the average hours students report working, and the rate at which their classes reach capacity.

To explore this, we analyzed data from the UCSD Historical Enrollment dataset and UCSD CAPEs. These datasets provide valuable information on class enrollment patterns, average grades, and the workload associated with different professors and courses. Through our analysis, we identified a consistent trend: professors who give lower average grades and require higher average hours of work tend to have classes that fill more slowly. Conversely, professors with higher average grades and lower reported hours of work see their classes fill at a much faster rate. These findings suggest that students are heavily influenced by factors like perceived workload and grades recieved when choosing courses.

Our work not only supports the idea that students prioritize "easier" professors but also highlights the complexity of enrollment decisions. Factors such as course requirements, scheduling constraints, and course reputation may also play a role in shaping enrollment rates. By understanding these dynamics, we can better anticipate enrollment patterns, providing insights for academic advisors and the UCSD student population. In conclusion, this analysis aims to highlight the importance of data science approache to understanding student behavior and improving the course selection process at UCSD.

How do factors from a professors historical data such as average hours worked and average grade received, impact the number of days and hours until a course at UCSD reaches the max capacity of enrolled students?



## Background and Prior Work

A stressful time that most, if not all UCSD students encounter is the time period of  having to enroll for classes. UCSD itself is a large university with currently over 45,000 enrolled, making enrollment competitive especially in popular majors like Biology or Computer Science. Many are faced with difficult decisions to make on whether or not they should first or second pass a class, worried that although they might enroll into one class, they might not get into another. The first pass allows students to enroll in a limited number of unit classes, allowing them to secure spots in high-demand classes. With the second pass, students get to complete their full schedule for the quarter, however some may run into classes that are already full. This puts them into a tricky dilemma: should they secure a spot in a safe class or gamble on waiting for a more popular class in the second pass?

With this, they might reach out to online resources such as SETS, CAPEs, or even the UCSD reddit itself. Through these platforms, they have a variety of statistical options to look at which a professors average grade received, average hours worked in the class, distribution of grades within the class itself, and as well as the expected grade. On top of the statistical options, they may also receive anecdotal advice from students themselves through reddit posts or asking friends who may have already taken the class. Considering these stressful circumstances and factors in play, they formulate a decision on whether or not to first or second pass a class.

# Hypothesis


We predict that there is a strong correlation between factors such as average grades received, average weekly study hours, and the enrollment rate of a class. Courses with lower average grades and higher weekly study hours tend to have a slower enrollment rate for that course instance, as these factors often indicate a perception of course satisfaction and manageability. Students are more likely to enroll in courses they believe will help them achieve a good grade or pass with less effort.

In short, we believe that the class having lower average grades, and requiring higher weekly study hours will contribute to slower enrollment rates.

Since there are more than a thousand unique course offerings at UCSD, we will only test this hypothesis within the Cognitive Science department at UCSD. We will analyze 2 types of Cognitive Science classes (required upper division, and a required lower division class). The term "fast enrollment rates" will be measured based on the number of days and hours until the maximum capacity of a class is reached for enrollment.

We can filter the CAPES dataset to see only professors that have taught COGS 107A.

```python
cogs_107a_prof = ucsd_capes.copy()

# filter to only cogs 107a course instances
cogs_107a_prof = cogs_107a_prof[cogs_107a_prof['Course'] == 'COGS 107A - Neuroanatomy and Physiology (A)']

# array to store unique profs
professors = []
for professor in cogs_107a_prof['Instructor']:
    if professor in professors:
        continue
    else:
        professors.append(professor)
print(professors)
```

These are all the professors who have taught COGS 107A. Lets find and take a look at their average study hours worked.

```python
# Find averages for each professor
profBoyle = cogs_107a_prof[(cogs_107a_prof['Course'] == 'COGS 107A - Neuroanatomy and Physiology (A)') & 
                                    (cogs_107a_prof['Instructor'] == 'Boyle Mary E. T.')]
profRangel = cogs_107a_prof[(cogs_107a_prof['Course'] == 'COGS 107A - Neuroanatomy and Physiology (A)') & 
                                    (cogs_107a_prof['Instructor'] == 'Rangel Lara M')]
profAlfonso = cogs_107a_prof[(cogs_107a_prof['Course'] == 'COGS 107A - Neuroanatomy and Physiology (A)') & 
                                    (cogs_107a_prof['Instructor'] == 'Alfonso Stephanie Isabel')]
profGreene = cogs_107a_prof[(cogs_107a_prof['Course'] == 'COGS 107A - Neuroanatomy and Physiology (A)') & 
                                    (cogs_107a_prof['Instructor'] == 'Greene Deanna Jacquelyn')]
profPineda = cogs_107a_prof[(cogs_107a_prof['Course'] == 'COGS 107A - Neuroanatomy and Physiology (A)') & 
                                    (cogs_107a_prof['Instructor'] == 'Pineda Jaime A')]

# Create 2 individual dataframes
professors = ['Rangel Lara M', 'Boyle Mary E. T.', 'Alfonso Stephanie Isabel', 'Greene Deanna Jacquelyn', 'Pineda Jaime A']
average_hours = [profRangel['Study Hours per Week'].mean(),
                profBoyle['Study Hours per Week'].mean(),
                 profAlfonso['Study Hours per Week'].mean(),
                   profGreene['Study Hours per Week'].mean(),
                   profPineda['Study Hours per Week'].mean(),
                ]

# combine the 2 dataframes
data = pd.DataFrame({
    'Professor': professors,
    'Average Study Hours': average_hours
})

df = pd.DataFrame(data)

plt.figure(figsize=(15, 6))
sns.barplot(x='Professor', y='Average Study Hours', data=df)
plt.xlabel('Professor')
plt.ylabel('Average Study Hours')
plt.title('Comparison of Average Study Hours by Professor For COGS 107A') 
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Lets now take a look at the average GPA students recieved in their classes.

```python
Boyle_data = ucsd_capes[(ucsd_capes['Instructor'] == 'Boyle Mary E. T.')]
Rangel_data = ucsd_capes[(ucsd_capes['Instructor'] == 'Rangel Lara M')]

# Function to convert letter grades to numeric values (only the numeric part)
def numeric_coversion(grade):
    # Numeric Values
    number_grade = {
        'A+': 4.0, 'A': 4.0, 'A-': 3.7,
        'B+': 3.3, 'B': 3.0, 'B-': 2.7,
        'C+': 2.3, 'C': 2.0, 'C-': 1.7,
        'D+': 1.3, 'D': 1.0, 'D-': 0.7,
        'F': 0.0}

    # Changes the grade format to only numeric values
    if isinstance(grade, str):
        grade = grade.split('(')[0].strip()  
        if grade in number_grade:
            return number_grade[grade]
    try:
        return float(grade)
    except ValueError:
        return None  


#Filtering the 'Average Grade Received' results into numeric values
Boyle_data.loc[:, 'Average Grade Received'] = Boyle_data['Average Grade Received'].apply(numeric_coversion)
Rangel_data.loc[:, 'Average Grade Received'] = Rangel_data['Average Grade Received'].apply(numeric_coversion)


# Plotting the data
plt.figure(figsize=(9, 7))

#Plotting Professor Boyle
sns.barplot(data=Boyle_data, x='Instructor', y='Average Grade Received')

#Plotting Professor Rangel
sns.barplot(data=Rangel_data, x='Instructor', y='Average Grade Received')

plt.title('Average Grade Received in COGS 107A', fontsize=14)
plt.xlabel('Course Professors', fontsize=12)
plt.ylabel('Average Grade', fontsize=12)
plt.tight_layout()
plt.show()
```

Looking at our other independent variable that were considering, which was the average grade recieved across Professor Boyle's COGS 107A course instance, we can also see that the average student GPA was lower compared to Professor Rangel. We can infer that this may have been another factor into why Professor Boyle's course instance for COGS 107A did not fill up.

In conclusion, after analyzing Professor Boyle's COGS 107A course instance in comparison with Professor Rangel's COGS 107A course instance, we can see that Boyle having a higher average of hours spent studying for the class as well as students recieiving a lower GPA/grade may have affected the result of of her class not filling up.

```python
Hoffman_data = ucsd_capes[(ucsd_capes['Instructor'] == 'Hoffman Drew Ellen')]
Barrera_data = ucsd_capes[(ucsd_capes['Instructor'] == 'Barrera Steven James')]

# Function to convert letter grades to numeric values (only the numeric part)
def numeric_coversion(grade):
    # Numeric Values
    number_grade = {
        'A+': 4.0, 'A': 4.0, 'A-': 3.7,
        'B+': 3.3, 'B': 3.0, 'B-': 2.7,
        'C+': 2.3, 'C': 2.0, 'C-': 1.7,
        'D+': 1.3, 'D': 1.0, 'D-': 0.7,
        'F': 0.0}

    # Changes the grade format to only numeric values
    if isinstance(grade, str):
        grade = grade.split('(')[0].strip()  
        if grade in number_grade:
            return number_grade[grade]
    try:
        return float(grade)
    except ValueError:
        return None  


#Filtering the 'Average Grade Received' results into numeric values
Hoffman_data.loc[:, 'Average Grade Received'] = Hoffman_data['Average Grade Received'].apply(numeric_coversion)
Barrera_data.loc[:, 'Average Grade Received'] = Barrera_data['Average Grade Received'].apply(numeric_coversion)


# Plotting the data
plt.figure(figsize=(9, 7))

#Plotting Professor Boyle
sns.barplot(data=Hoffman_data, x='Instructor', y='Average Grade Received')

#Plotting Professor Rangel
sns.barplot(data=Barrera_data, x='Instructor', y='Average Grade Received')

plt.title('Average Grade Received in COGS 14B', fontsize=14)
plt.xlabel('Course Professors', fontsize=12)
plt.ylabel('Average Grade', fontsize=12)
plt.tight_layout()
plt.show()
```

Although Hoffman had the higher average of study hours spent on her iteratation of COGS 14B, professor Barrera's COGS 14B classes had a lower GPA/Grade distribution among students.

Based off this we cannot really infer if the average grade recieved within the classes may have played an effect in how quickly the class filled up.

In conclusion, after analyzing Professor Hoffman's COGS 14B course instance in comparison with Professor Barrera's COGS 14B course instance, we can see that Hoffman having a higher average of hours spent studying for the class may have affected the result of the time it took for the class to fill up. On the other hand though, her classes had a higher grade/gpa distribution compared to Professor Barrera, which may infer that grades were not a big enough factor in students enrolling into the course instance.
